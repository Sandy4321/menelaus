```python
#TODO: this is a bit of a hack, to add /molten_export/ to sys.path, so we can export molten while within the repo
#probably want to do this in a cleaner way before public release
import os, sys
sys.path.insert(0, os.path.abspath(os.path.join(os.path.dirname('Testing_Vignette.ipynb'), '..', '..')))

from molten.distribution.PCA_CD import PCA_CD
# from src.drift_metrics import drift_metrics
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
d = 2
step_size = 2

# data = pd.read_csv(f'tests/SyntheticData/sim_data_{d}d.csv')[::step_size]
data = pd.read_csv(f'S:/User Workspace/tschill/molten/tests/SyntheticData/sim_data_{d}d.csv')[::step_size]
data.index = range(len(data))

normalized_data = StandardScaler().fit_transform(data)
data = pd.DataFrame(normalized_data, index=data.index)

window_size = 250
verbose = True

################################################
###################### KL ######################
################################################
divergence_metric = "KL"
pca_cd_KL = PCA_CD(window_size=window_size, verbose=verbose, divergence_metric=divergence_metric)
pca_cd_KL.fit(data=data)
pca_cd_KL_df = pca_cd_KL.get_drift_tracker()

# pca_cd_KL_df.to_csv(f"tests/SyntheticData/Data Distribution/data/pca_cd_{d}d_KL.csv", index=False)
```

```python
pca_cd_KL_df.index = [x*step_size for x in pca_cd_KL_df.index]
pca_cd_KL_df.ids = [x*step_size for x in pca_cd_KL_df.ids]

#### Compute Metrics ####
true_drift_inds = [5000, 10000, 15000, 20000]
pred_drift_inds = [x for x, y in zip(pca_cd_KL_df.ids, pca_cd_KL_df.drift_detected) if y]

pca_cd_KL_metrics = drift_metrics(true_drift_inds=true_drift_inds, pred_drift_inds=pred_drift_inds,
                                  window_size=window_size)

#### Build Plot ####
with plt.style.context('fivethirtyeight'):
    plt.figure(figsize=(20, 8))
    plt.plot(pca_cd_KL_df["ids"], pca_cd_KL_df["page_hinkley_differences"], linestyle='-', marker=',', zorder=0, linewidth=.85)
    plt.plot(pca_cd_KL_df["ids"], pca_cd_KL_df["theta_threshold"], linestyle='--', marker=',', color="goldenrod", zorder=1, linewidth=1)
    plt.vlines(x=pca_cd_KL_df.loc[pca_cd_KL_df.drift_detected == True, "ids"],
               ymin=0, ymax=pca_cd_KL_df.loc[pca_cd_KL_df.drift_detected == True, "theta_threshold"],
               linestyle='-', color="darkred", zorder=2, linewidth=5, alpha=0.25)
    plt.title("Kullback-Leibler Divergence", fontsize=16, y=1.01)
    plt.suptitle("PCA Change Detection", fontsize=24, y=0.97)
    plt.xlabel("Time")
    plt.ylabel("Divergence Value")

    ax = plt.gca()
    ax.axes.yaxis.set_ticks([])
    plt.savefig(f"tests/SyntheticData/Data Distribution/visuals/pca_cd_{d}d_KL.png")


################################################
################# Intersection #################
################################################
divergence_metric = "Intersection"
pca_cd_INT = PCA_CD(window_size=window_size, verbose=verbose, divergence_metric=divergence_metric)
pca_cd_INT.fit(data=data)
pca_cd_INT_df = pca_cd_INT.get_drift_tracker()

pca_cd_INT_df.to_csv(f"tests/SyntheticData/Data Distribution/data/pca_cd_{d}d_INT.csv", index=False)

pca_cd_INT_df.index = [x*step_size for x in pca_cd_INT_df.index]
pca_cd_INT_df.ids = [x*step_size for x in pca_cd_INT_df.ids]

#### Compute Metrics ####
true_drift_inds = [5000, 10000, 15000, 20000]
pred_drift_inds = [x for x, y in zip(pca_cd_INT_df.ids, pca_cd_INT_df.drift_detected) if y]

pca_cd_INT_metrics = drift_metrics(true_drift_inds=true_drift_inds, pred_drift_inds=pred_drift_inds,
                                   window_size=window_size)

#### Build Plot ####
with plt.style.context('fivethirtyeight'):
    plt.figure(figsize=(20, 8))
    plt.plot(pca_cd_INT_df["ids"], pca_cd_INT_df["page_hinkley_differences"],
             linestyle='-', marker=',', zorder=0, linewidth=.85)
    plt.plot(pca_cd_INT_df["ids"], pca_cd_INT_df["theta_threshold"],
             linestyle='--', marker=',', color="goldenrod", zorder=1, linewidth=1)
    plt.vlines(x=pca_cd_INT_df.loc[pca_cd_INT_df.drift_detected == True, "ids"],
               ymin=0, ymax=pca_cd_INT_df.loc[pca_cd_INT_df.drift_detected == True, "theta_threshold"],
               linestyle='-', color="darkred", zorder=2, linewidth=5, alpha=0.25)
    plt.title("Intersection Divergence", fontsize=16, y=1.01)
    plt.suptitle("PCA Change Detection", fontsize=24, y=0.97)
    plt.xlabel("Time")
    plt.ylabel("Divergence Value")

    ax = plt.gca()
    ax.axes.yaxis.set_ticks([])
    plt.savefig(f"tests/SyntheticData/Data Distribution/visuals/pca_cd_{d}d_INT.png")

################################################
###################### LLH #####################
################################################
divergence_metric = "LLH"
pca_cd_LLH = PCA_CD(window_size=window_size, verbose=verbose, divergence_metric=divergence_metric)
pca_cd_LLH.fit(data=data)
pca_cd_LLH_df = pca_cd_LLH.get_drift_tracker()

pca_cd_LLH_df.to_csv(f"tests/SyntheticData/Data Distribution/data/pca_cd_{d}d_LLH.csv", index=False)

pca_cd_LLH_df.index = [x*step_size for x in pca_cd_LLH_df.index]
pca_cd_LLH_df.ids = [x*step_size for x in pca_cd_LLH_df.ids]

#### Compute Metrics ####
true_drift_inds = [5000, 10000, 15000, 20000]
pred_drift_inds = [x for x, y in zip(pca_cd_LLH_df.ids, pca_cd_LLH_df.drift_detected) if y]

pca_cd_LLH_metrics = drift_metrics(true_drift_inds=true_drift_inds, pred_drift_inds=pred_drift_inds,
                                   window_size=window_size)

#### Build Plot ####
with plt.style.context('fivethirtyeight'):
    plt.figure(figsize=(20, 8))
    plt.plot(pca_cd_LLH_df["ids"], pca_cd_LLH_df["page_hinkley_differences"],
             linestyle='-', marker=',', zorder=0, linewidth=.85)
    plt.plot(pca_cd_LLH_df["ids"], pca_cd_LLH_df["theta_threshold"],
             linestyle='--', marker=',', color="goldenrod", zorder=1, linewidth=1)
    plt.vlines(x=pca_cd_LLH_df.loc[pca_cd_LLH_df.drift_detected == True, "ids"],
               ymin=0, ymax=pca_cd_LLH_df.loc[pca_cd_LLH_df.drift_detected == True, "theta_threshold"],
               linestyle='-', color="darkred", zorder=2, linewidth=5, alpha=0.25)
    plt.title("Log-Likelihood Divergence", fontsize=16, y=1.01)
    plt.suptitle("PCA Change Detection", fontsize=24, y=0.97)
    plt.xlabel("Time")
    plt.ylabel("Divergence Value")

    ax = plt.gca()
    ax.axes.yaxis.set_ticks([])
    plt.savefig(f"tests/SyntheticData/Data Distribution/visuals/pca_cd_{d}d_LLH.png")
```
